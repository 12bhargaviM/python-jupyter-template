### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt

try:
    x = range(0, 11)
    y = [i**2 if isinstance(i, (int, float)) else 0 for i in x]
    
    if len(x) != len(y):
        raise ValueError("x and y must have equal lengths")
        
    plt.figure(figsize=(8, 5))
    plt.plot(x, y, 'g--o', linewidth=2)
    plt.title("Square Function with Validation", pad=20)
    plt.xlabel("X Values")
    plt.ylabel("Y = X²")
    plt.grid(True, alpha=0.3)
    plt.show()
    
except Exception as e:
    print(f"Error: {str(e)}")
```
