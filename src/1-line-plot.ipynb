### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
```

```python
import matplotlib.pyplot as plt

# Generate data
x = range(0, 11)  # Values from 0 to 10
y = [i**2 for i in x]  # Squares of x values

# Create the plot
plt.figure(figsize=(8, 5))  # Optional: Set figure size
plt.plot(x, y, 'g--', linewidth=2, markersize=8, marker='o')  # Green dashed line with circle markers

# Customize the plot
plt.title("Square Function Visualization", fontsize=14, pad=20)
plt.xlabel("X Values", fontsize=12)
plt.ylabel("Square of X", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)  # Semi-transparent dashed grid

# Display the plot
plt.show()
```
