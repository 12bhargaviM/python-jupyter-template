### Exercise 2: Bar Plot with Colors

```python
# TASK: Create a bar plot with the following data: categories = ['A', 'B', 'C', 'D'] and values = [5, 7, 3, 9].
# Use different colors for each bar and add a title to the plot.
```

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [15, 25, 30]

try:
    if len(categories) != len(values):
        raise ValueError("Categories and values must have equal lengths")
    if not all(isinstance(v, (int, float)) for v in values):
        raise TypeError("All values must be numeric")
    
    colors = ['#FF6B6B', '#4ECDC4', '#45B7D1']  # Modern color palette
    plt.bar(categories, values, color=colors, width=0.6)
    plt.title("Validated Bar Chart", pad=15)
    plt.ylim(0, max(values)*1.1)
    plt.show()

except Exception as e:
    print(f"Plotting Error: {str(e)}")
```
