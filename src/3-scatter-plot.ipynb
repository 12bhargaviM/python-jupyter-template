### Exercise 3: Scatter Plot with Annotations

```python
# TASK: Create a scatter plot with x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Annotate each point with its (x, y) value, and set the title as 'Scatter Plot with Annotations'.
```

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

try:
    if len(x) != len(y):
        raise ValueError("x and y coordinates count mismatch")
    if not all(isinstance(i, (int, float)) for pair in zip(x,y) for i in pair):
        raise TypeError("All coordinates must be numeric")
    
    plt.figure(figsize=(7,7))
    scatter = plt.scatter(x, y, c='#9C27B0', s=100, alpha=0.7)
    
    # Annotations with adjusted positions
    for i, (xi, yi) in enumerate(zip(x, y)):
        offset = (0, 15) if i%2 else (0, -20)  # Alternate label positions
        plt.annotate(f'({xi},{yi})', (xi, yi), 
                    xytext=offset, textcoords='offset points',
                    ha='center', fontsize=9)
    
    plt.title("Validated Scatter Plot", pad=15)
    plt.grid(True, linestyle=':', alpha=0.4)
    plt.show()

except Exception as e:
    print(f"Scatter Plot Error: {str(e)}")
```
