### Exercise 4: Pie Chart with Percentages

```python
# TASK: Create a pie chart with the following data: labels = ['Python', 'Java', 'C++', 'JavaScript'] and sizes = [40, 25, 20, 15].
# Display the percentages on the chart using autopct.
```

```python
import matplotlib.pyplot as plt

labels = ['Python', 'Java', 'C++', 'JavaScript']
sizes = [40, 25, 20, 15]

try:
    if abs(sum(sizes) - 100) > 0.01:  # Account for floating point precision
        sizes = [s/sum(sizes)*100 for s in sizes]  # Normalize to 100%
    
    if not all(isinstance(s, (int, float)) for s in sizes):
        raise TypeError("Size values must be numeric")
    
    explode = (0.1, 0, 0, 0)
    colors = ['#FFC107', '#2196F3', '#4CAF50', '#E91E63']
    
    plt.pie(sizes, explode=explode, labels=labels, colors=colors,
            autopct='%1.1f%%', shadow=True, startangle=140,
            textprops={'fontsize': 9})
    plt.title("Validated Language Popularity", pad=20)
    plt.tight_layout()
    plt.show()

except Exception as e:
    print(f"Pie Chart Error: {str(e)}")
```
