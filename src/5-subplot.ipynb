### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

try:
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle('Validated Subplot Dashboard', y=1.02, fontsize=14)
    
    # Line Plot
    x_line = np.linspace(0, 10, 100)
    axs[0,0].plot(x_line, np.sin(x_line), 'r-')
    axs[0,0].set_title('Sine Wave')
    
    # Bar Plot
    categories = ['A', 'B', 'C']
    values = [15, 25, 30]
    axs[0,1].bar(categories, values, color=['teal', 'coral', 'indigo'])
    axs[0,1].set_title('Bar Chart')
    
    # Scatter Plot
    x_scatter = np.random.rand(50)
    y_scatter = np.random.rand(50)
    axs[1,0].scatter(x_scatter, y_scatter, c='green', alpha=0.6)
    axs[1,0].set_title('Scatter Plot')
    
    # Pie Chart
    sizes = [35, 25, 40]
    axs[1,1].pie(sizes, labels=['X','Y','Z'], autopct='%1.1f%%',
                colors=['gold','lightcoral','lightblue'])
    axs[1,1].set_title('Pie Chart')
    
    plt.tight_layout(pad=3.0)
    plt.show()

except Exception as e:
    print(f"Subplot Error: {str(e)}")
```
