### Exercise 5: Subplot Layout

```python
# TASK: Create a 2x2 subplot layout.
# Plot a line chart in the first subplot, a bar chart in the second, a scatter plot in the third, and a pie chart in the fourth.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure and subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
fig.suptitle('2x2 Subplot Example', fontsize=16)

# Line chart (top-left)
x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x), 'r-')
axs[0, 0].set_title('Sine Wave')

# Bar chart (top-right)
categories = ['A', 'B', 'C']
values = [15, 25, 30]
axs[0, 1].bar(categories, values, color=['blue', 'green', 'red'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot (bottom-left)
x = np.random.rand(50)
y = np.random.rand(50)
axs[1, 0].scatter(x, y, c='purple', alpha=0.6)
axs[1, 0].set_title('Scatter Plot')

# Pie chart (bottom-right)
labels = ['X', 'Y', 'Z']
sizes = [35, 25, 40]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

# Adjust layout
plt.tight_layout()
plt.show()
```
