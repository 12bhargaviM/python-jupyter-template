### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)  # Sample data

try:
    if not all(isinstance(x, (int, float)) for x in data):
        raise TypeError("Histogram data must be numeric")
    
    plt.figure(figsize=(9, 6))
    plt.hist(data, bins=30, color='#607D8B', 
             edgecolor='#455A64', alpha=0.7,
             density=True)  # Normalized histogram
    
    plt.title("Validated Normal Distribution", pad=15)
    plt.xlabel("Value")
    plt.ylabel("Density")
    plt.grid(axis='y', alpha=0.3)
    plt.show()

except Exception as e:
    print(f"Histogram Error: {str(e)}")
```
