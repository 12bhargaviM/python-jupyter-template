### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
plt.suptitle('2x2 Subplot Demonstration')

# Line plot (top-left)
x = np.linspace(0, 10, 100)
axs[0, 0].plot(x, np.sin(x), 'b-')
axs[0, 0].set_title('Sine Wave')

# Bar plot (top-right)
categories = ['A', 'B', 'C']
values = [15, 25, 30]
axs[0, 1].bar(categories, values, color=['red', 'green', 'blue'])
axs[0, 1].set_title('Bar Chart')

# Scatter plot (bottom-left)
x = np.random.rand(50)
y = np.random.rand(50)
axs[1, 0].scatter(x, y, c='purple', alpha=0.6)
axs[1, 0].set_title('Scatter Plot')

# Pie chart (bottom-right)
labels = ['X', 'Y', 'Z']
sizes = [35, 25, 40]
axs[1, 1].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axs[1, 1].set_title('Pie Chart')

plt.tight_layout()
plt.show()
```
